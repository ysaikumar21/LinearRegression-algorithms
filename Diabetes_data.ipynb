```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn import datasets
```

```python
df = datasets.load_diabetes()
```

```python
dir(df)
```

```python
import pandas as pd
```

```python
f = pd.DataFrame(data = df['data'],columns=df['feature_names'])
f['dep'] = df['target']
f
```

```python
y=f['dep']
X=f.drop('dep',axis=1)
print(X.head())
print(y.head())
```

```python
X_train,X_test,y_train,y_test = sklearn.model_selection.train_test_split(X,y,test_size=.3,random_state=1)
X_train
```

```python
y_train
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
y_train_pred=lm.predict(X_train)
y_test_pred=lm.predict(X_test)
```

```python
sf=pd.DataFrame([[y_train,y_train_pred]])
sf.head()
```

```python
mse=mean_squared_error(y_test,y_test_pred)
print(mse)
```

```python
plt.scatter(y_train,y_train_pred,marker='o',color='blue',label='Training data')
plt.scatter(y_test,y_test_pred,marker='*',color='red',label='Testing data')
plt.xlabel('Predicted values')
plt.ylabel('True values')
plt.legend(loc='upper left')
plt.plot()
plt.show()
```

```python
#Accuracy
from sklearn.metrics import r2_score
```

```python
#training data Accuracy
r2_score(y_train,y_train_pred)
```

```python
#Testing data Accuracy
r2_score(y_test,y_test_pred)
```

**Training Data**

'Actual Values and Predicted values and differences blw the two variables of actual and predicted and Accuracy %'

```python
df=pd.DataFrame()
df['y_train']=y_train
df['y_train_pred']=y_train_pred
df['Loss']=y_train-y_train_pred
df['Accuracy in percentage %']=r2_score(y_train,y_train_pred)*100
df
```

**Testing Data**

'Actual value and predicted values and loss and Accuracy %'

```python
df=pd.DataFrame()
df['y_test']=y_test
df['y_test_pred']=y_test_pred
df['Loss']=y_test-y_test_pred
df['Accuracy in percentage %']=r2_score(y_test,y_test_pred)*100
df
```
