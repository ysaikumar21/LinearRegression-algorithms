```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
myfile=pd.DataFrame(pd.read_csv('Simple_dataset.csv'))
```

```python
myfile
```

```python
X=myfile.drop('Salary',axis=1)
y=myfile['Salary']
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =sklearn.model_selection.train_test_split(X,y,test_size=0.7,random_state=11)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

**Training Data**

```python
y_train_pred=lm.predict(X_train)
y_train_pred
```

```python
df=pd.DataFrame()
df['YearsExperience']=X_train
df['Salary']=y_train
df['Predicted Salary']=y_train_pred
df
```

```python
df=pd.DataFrame()
df['YearsExperience']=X_train
df['Salary']=y_train
df['Predicted Salary']=y_train_pred
df['Loss']=y_train-y_train_pred
df
```

```python
from sklearn.metrics import r2_score
```

```python
#Accuracy
r2_score=(y_train,y_train_pred)
```

```python
mse=mean_squared_error(y_train,y_train_pred)
mse
```

```python
plt.figure(figsize=(5,3))
plt.title('Training Data')
plt.scatter(X_train,y_train,color='blue',marker='o')
plt.plot(X_train,y_train_pred,color='y',marker='*')
plt.xlabel('YearsExperiences')
plt.ylabel('Salary')
plt.show()
```

**Testing Data**

```python
y_test_pred=lm.predict(X_test)
y_test_pred
```

```python
df=pd.DataFrame()
df['YearsExperience']=X_test
df['Salary']=y_test
df['Predicted Salary']=y_test_pred
df['Loss']=y_test-y_test_pred
df
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_test_pred)
```

```python
mse=mean_squared_error(y_test,y_test_pred)
mse
```

```python
plt.figure(figsize=(5,3))
plt.title('Testing Data')
plt.scatter(X_test,y_test,color='blue',marker='o')
plt.plot(X_test,y_test_pred,color='y',marker='*')
plt.xlabel('YearsExperiences')
plt.ylabel('Salary')
plt.show()
```
